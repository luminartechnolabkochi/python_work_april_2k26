### FILE READ AND WRITE

```python
fr = open("C:\\Users\\Luminar\\Desktop\\python_works\\fileoperations\\regional_movies_comma.txt","r")

fw_malayalam=open("C:\\Users\\Luminar\\Desktop\\python_works\\fileoperations\\malayalam_movies.txt","w")

fw_kannada=open("C:\\Users\\Luminar\\Desktop\\python_works\\fileoperations\\kannada_movies.txt","w")

for line in fr:#line=Drishyam, Malayalam\n

    line = line.rstrip("\n")#line="Drishyam, Malayalam"

    title,language=line.split(",")#["Drishyam","Malayalam"]

    if language==" Malayalam":

        fw_malayalam.write(title+"\n")

    else:

        fw_kannada.write(title+"\n")

fw_malayalam.close()
fw_kannada.close()






```

```python
fr_items=open("C:\\Users\\Luminar\\Desktop\\python_works\\fileoperations\\restaurant_menu_dataset.txt","r")

fw_veg=open("C:\\Users\\Luminar\\Desktop\\python_works\\fileoperations\\veg-items.txt","w")

fw_nonveg=open("C:\\Users\\Luminar\\Desktop\\python_works\\fileoperations\\nonveg-items.txt","w")


for line in fr_items:#line="Paneer Butter Masala, Veg\n"

    line = line.rstrip("\n")##line="Paneer Butter Masala, Veg"

    item,category=line.split(",")

    category=category.lstrip(" ")

  

    if category == "Veg":

        fw_veg.write(item+"\n")

    else:

        fw_nonveg.write(item+"\n")


fw_veg.close()

fw_nonveg.close()

```
