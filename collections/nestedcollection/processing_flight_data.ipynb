```python
fr = open("C:\\Users\\Luminar\\Desktop\\python_works\\collections\\nestedcollection\\flights.csv")
```

```python
flights=[]

fr.readline()

for line in fr:

    line = line.rstrip("\n")

    year,month,passengers=line.split(",")

    dictionary = {"year":year,"month":month,"passengers":passengers}
    
    flights.append(dictionary)
print(flights)

```

```python

result = {}

for records in flights:

    year = records["year"]

    passenger_count = int(records["passengers"])

    if year in result:

        result[year]+=passenger_count
    else:

        result[year]=passenger_count
print(result)
```

```python

import matplotlib.pyplot as plt

plt.figure(figsize=(12,8))

x=list(result.keys())
y=list(result.values())

plt.bar(x,y)
plt.xlabel("year")
plt.ylabel("passenger count")

plt.show()
```
