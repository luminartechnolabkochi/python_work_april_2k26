```python

fifa_world_cup = [
# [Country, Wins, Appearances]
    ["Germany", 4, 21],
    ["Spain", 1, 17],
    ["Italy", 4, 18],
    ["Brazil", 5, 23],
    ["Netherlands", 0, 10],
    ["Argentina", 3, 19],
    ["Uruguay", 2, 15],
    ["France", 2, 17],
    ["England", 1, 17],
    ["Mexico", 0, 18]
]
```

### sort fifa_world_cup wrt appearances order by desc

```python
sorted_data=sorted(fifa_world_cup,key=lambda lst:lst[-1],reverse=True )

sorted_data
```

```python

teams= [record[0] for record in sorted_data ]
appearances=[record[-1] for record in sorted_data]

import matplotlib.pyplot as plt
plt.figure(figsize=(12,8))
plt.bar(teams,appearances)
plt.show()
```

### sort fifa_world_cup wrt wins order by desc

```python
srt_wins=sorted(fifa_world_cup,key=lambda lst:lst[1],reverse=True)
print(srt_wins)
```

```python
most_win_team= srt_wins[0]

print(most_win_team)
```

```python
least_win_team = srt_wins[-1]
print(least_win_team)
```

### is there any team with 10 world cup apperances

```python

is_all_team_appeared_10 =[team[-1]==10 for team in fifa_world_cup]

is_appeard_10 = any(is_all_team_appeared_10)

print(is_appeard_10)
```

### is all team played atleast 5 matches ?

```python
all_team_played_atleast_5 = [team[-1]>=5 for team in fifa_world_cup]

all(all_team_played_atleast_5)
```
